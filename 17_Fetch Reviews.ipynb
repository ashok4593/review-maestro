```python
from google_play_scraper import Sort, reviews
review, continuation_token = reviews(
    'us.zoom.videomeetings',
    count = 100
)
for i in range(1,10):
    result, continuation_token = reviews(
        'us.zoom.videomeetings',
        count = 100,
        continuation_token=continuation_token # defaults to None(load from the beginning)
    )
    review.extend(result)
reviews_dataset = pd.DataFrame(data = review).loc[:,['reviewId','userName','content','score','at']] 
```

```python
reviews_dataset.rename(columns = {'content':'review','userName':'authorName','at':'postedAt','score':'starRating'}, inplace = True)
```
